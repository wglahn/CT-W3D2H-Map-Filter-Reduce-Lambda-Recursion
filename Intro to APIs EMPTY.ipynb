```python
def get_racer_info(data):
    new_data=[]
    for racer in data:
        racer_dict={}
        racer_name = f'{racer["Driver"]["givenName"]} {racer["Driver"]["familyName"]}'
        racer_dict[racer_name] = {
            "first_name": racer['Driver']['givenName'],
            "last_name": racer["Driver"]["familyName"],
            "postion": racer["position"],
            "wins": racer["wins"],
            "DOB": racer["Driver"]["dateOfBirth"],
            "nationality": racer["Driver"]["nationality"],
            "constructor": racer["Constructors"][0]["name"]
        }
        new_data.append(racer_dict)
    return new_data

print(get_racer_info(data))
```

Use the requests package to connect to this API and get and store data for 5 different pokemon.
Get the pokemons: name, atleast one ability's name, base_experience, and the URL for its sprite (an image that shows up on screen) for the 'front_shiny'

```python
import requests, json

# Gather information about 5 different Pokemon
# Name, 1 ability's name, base_experience, and the URL for its sprite (front_shiny) 

# estbalish pokemon name to get the correct link
# because all the links are set up with the same dictionary line up
# then we establish the route to each piece of information using a new dictionary
# return the new dictionary
# invoke the function 5 times with 5 different pokemon names

names = ['pikachu','slaking','golem','nosepass','arbok']

def get_poke_info(name):
    url = f"https://pokeapi.co/api/v2/pokemon/{name}"
    response = requests.get(url)
    if response.ok:
        poke_dict={
            "name":response.json()['forms'][0]['name'],
            "ability":response.json()['abilities'][0]['ability']['name'],
            "base xp":response.json()['base_experience'],
            "sprite": response.json()['sprites']['front_shiny']
            }
        return poke_dict
    else:
        print("Incorrect pokemon name. Please try again.")
        
poke_info = []
for name in names:
    poke_info.append(get_poke_info(name))
    
print(poke_info)
```
